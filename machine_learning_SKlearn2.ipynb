```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
data = pd.read_csv(uri)
data.head()
```

```python
data.shape
```

```python
x = data[["home", "how_it_works", "contact"]]
y = data["bought"]

y.head()
```

```python
x.head()
```

```python
#train 75% data
train_x = x[:75]
train_y = y[:75]
```

```python
#test 25% data
test_x = x[75:]
test_y = y[75:]
```

```python
print('we will train with %d elements and test with %d elements' % (len(train_x), len(test_x)))
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

model = LinearSVC()
model.fit(train_x, train_y)
predictions = model.predict(test_x)

accuracy = accuracy_score(test_y, predictions) * 100
print('the accuracy was %.2f%%' % accuracy)
```

# Using library Train Test Split

```python
from sklearn.model_selection import train_test_split

SEED = 20
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state=SEED, test_size = 0.25, stratify=y)
print('we will train with %d elements and test with %d elements' % (len(train_x), len(test_x)))

model = LinearSVC()
model.fit(train_x, train_y)
predictions = model.predict(test_x)

accuracy = accuracy_score(test_y, predictions) * 100
print('the accuracy was %.2f%%' % accuracy)
```
